### **Q5. Há *outliers* presentes?** / **Q5. Are *outliers* present?**

Pela distribuição do histograma, é possível verificar indícios da presença de *outliers*. Olhe por exemplo as variáveis `price` e `minimum_nights`.

Os valores não seguem uma distribuição, e distorcem toda a representação gráfica. Para confirmar, há duas maneiras rápidas que auxiliam a detecção de *outliers*. São elas:

* Resumo estatístico por meio do método `describe()`
* Plotar `boxplots` para a variável.

--

Through the distribution of the histogram, it is possible to verify signs of the presence of *outliers*. Look for example at the variables `price` and `minimum_nights`.

The values ​​do not follow a distribution, and distort the entire graphical representation. To confirm, there are two quick ways that help detect *outliers*. Are they:

* Statistical summary using the `describe()` method
* Plot `boxplots` for the variable.

```
# ver o resumo estatístico das variáveis numéricas

df[['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month',
    'calculated_host_listings_count', 'availability_365', 'number_of_reviews_ltm']].describe()
```

Olhando o resumo estatístico acima, podemos confirmar algumas hipóteses como:

* A variável `price` possui 75% do valor abaixo de 24286, porém seu valor máximo é 999999.
* A quantidade mínima de noites (`minimum_nights`) está em 365 dias ou seja, somente aluga por ano.

--

Looking at the statistical summary above, we can confirm some hypotheses such as:

* The `price` variable has 75% of the value below 24286, but its maximum value is 999999.
* The minimum number of nights (`minimum_nights`) is 365 days, i.e., only rent per year.

#### Boxplot para minimum_nights / Boxplot to minimum_nights


```
# minimum_nights

df.minimum_nights.plot(kind='box', vert=False, figsize=(15, 3))
plt.show()

# Mostrar quantidade de valores acima de 30 dias para minimum_nights
# Show number of values ​​over 30 days for minimum_nights

print("minimum_nights: \nValores acima de 30:")
print("{} entradas".format(len(df[df.minimum_nights > 30])))
print("{:.4f}%".format((len(df[df.minimum_nights > 30]) / df.shape[0])*100))
```

#### Boxplot para price / Boxplot to price

Note que estamos lidando com valores em Ienes, ou seja, devido ao câmbio que estamos acostumados no Brasil, pode parecer elevado, mas é comum  na economia japonesa.

Procuramos identificar uma quantidade de casos a partir de certo preço que represente próximo de 5% do número total, assim ficamos confortáveis em considerar valores acima como outliers.

--

Note that we are dealing with values ​​in Yen, that is, due to the exchange rate that are used in USA or in Europe, it may seem high, but it is common in the Japanese economy.

We seek to identify a number of cases from a certain price that represents close to 5% of the total number, so we are comfortable considering values ​​above as outliers.

```
# price

df.price.plot(kind='box', vert=False, figsize=(15, 3),)
plt.show()

# Mostrar quantidade de valores acima de 50000 para price
# Show number of values ​​above 50000 for price

print("\nprice: valores acima de 50000")
print("{} entradas".format(len(df[df.price > 50000])))
print("{:.4f}%".format((len(df[df.price > 50000]) / df.shape[0])*100))
```

#### Histogramas sem *outliers* / Histograms without *outliers*

Já que identificamos *outliers* nas variáveis `price` e `minimum_nights`, vamos agora limpar o *DataFrame* delas e plotar novamente o histograma.

Obs: Sobre a coluna `minimum_nights`. Embora alugar o imóvel por ano não seja exatamente um outlier (pode ser o formato que o host quis trabalhar), é interessante normalizar os dados de acordo com o comum para esse tipo de aluguel (no máximo, mensal).

--

Since we identified *outliers* in the `price` and `minimum_nights` variables, let's now clean their *DataFrame* and plot the histogram again.

Note: About the `minimum_nights` column. Although renting the property per year is not exactly an outlier (it may be the format the host wanted to work with), it is interesting to normalize the data according to what is common for this type of rental (maximum monthly).

```
# Remover os *outliers* em um novo DataFrame
# Remove *outliers* in a new DataFrame

df_clean = df.copy()
df_clean.drop(df_clean[df_clean.price > 50000].index, axis=0, inplace=True)
df_clean.drop(df_clean[df_clean.minimum_nights > 30].index, axis=0, inplace=True)

# Remover `neighbourhood_group`, pois está vazio
# Remove `neighborhood_group` as it is empty

df_clean.drop('neighbourhood_group', axis=1, inplace=True)

# Plotar o histograma para as variáveis numéricas
# Plot the histogram for numeric variables

df_clean.hist(bins=15, figsize=(15,10));
```

### **Q6. Qual a correlação existente entre as variáveis** / **Q6. What is the correlation between the variables**

Correlação significa que existe uma relação entre duas coisas. No nosso contexto, estamos buscando relação ou semelhança entre duas variáveis.

Essa relação pode ser medida, e é função do coeficiente de correlação estabelecer qual a intensidade dela. Para identificar as correlações existentes entre as variáveis de interesse, vou:

* Criar uma matriz de correlação
* Gerar um *heatmap* a partir dessa matriz, usando a biblioteca `seaborn`

--

Correlation means that there is a relationship between two things. In our context, we are looking for a relationship or similarity between two variables.

This relationship can be measured, and it is the function of the correlation coefficient to establish its intensity. To identify the correlations that exist between the variables of interest, I will:

* Create a correlation matrix
* Generate a *heatmap* from this matrix, using the `seaborn` library

```
# Criar uma matriz de correlação
# Create a correlation matrix

corr = df_clean[['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month',
    'calculated_host_listings_count', 'availability_365']].corr()

# Mostrar a matriz de correlação
# Show the correlation matrix

display(corr)
```

```
# plotar um heatmap a partir das correlações

sns.heatmap(corr, cmap='RdBu', fmt='.2f', square=True, linecolor='white', annot=True);
```

Lembrando que quanto mais próximo de 1, maior a correlação (mais azul).

--

Remembering that the closer to 1, the greater the correlation (more blue).

### **Q8. Qual a localidade mais cara do dataset?** / **Q8. What is the most expensive location in the dataset?**

Uma maneira de se verificar uma variável em função da outra é usando `groupby()`. No caso, queremos comparar os bairros (*neighbourhoods*) a partir do preço de locação.

--

One way to check one variable against another is using `groupby()`. In this case, we want to compare neighborhoods (*neighborhoods*) based on the rental price.

```
# Mostrar preços por bairros, na média
# Show prices by neighborhoods, on average

df_clean.groupby(['neighbourhood']).price.mean().sort_values(ascending=False)[:10].round(2)
```

As Latitudes and Longitudes are provided for properties, it is possible to plot each point. To do this, consider `x=longitude` and `y=latitude`.

```
# Plotar os imóveis pela latitude-longitude
# Plot properties by latitude-longitude

df_clean.plot(kind="scatter", x='longitude', y='latitude', alpha=0.4, c=df_clean['price'], s=8,
              cmap=plt.get_cmap('jet'), figsize=(12,8));
```

```
# Mostrar a média da coluna `minimum_nights`
# Show the average of the `minimum_nights` column

print("Média de mínimo de noites / Average minimum nights: {:.2f}".format(df_clean.minimum_nights.mean()))
```
